# 2.Examine the variables ‘Hardness’, ‘Sulfate’, and ‘Organic_carbon’ and comment on your observations.

```python
df.columns
```

```python
plt.figure()
plt.plot(df['ph'],df['Hardness'],label='Hardness')
plt.plot(df['ph'],df['Sulfate'],label='Sulfate')
plt.plot(df['ph'],df['Organic_carbon'],label='Organic_carbon')
plt.legend()
```

The amount of sulfate is high in water samples, followed by hardness and the water samples contains organic carbon as less amount.

```python
df[['Hardness','Sulfate','Organic_carbon']].plot(kind='hist')
```

The amount of organic carbon which contained in water samples are lies between 0 and 20, sulfate lies between 100 and 250 and hardness is in the range of 200 and 500.

```python
df[['Hardness','Sulfate','Organic_carbon']].mean()
```

```python
data = ['Hardness','Sulfate', 'Organic_carbon']
sns.pairplot(df[data])
```

```python
df[data].plot(kind = 'box')
```

the variables ‘Hardness’, ‘Sulfate’, and ‘Organic_carbon’ are normally distributed with outliers.From the pair plot, we can observe that there is no significant corelation between the 3 features.

# 4. Carry out correlation analysis on the dataset and find the correlation factor between

#### a. ph and Chloramines

#### b. ph and Conductivity What do we infer from this correlation value

```python
datac=df.corr()
datac
```

```python
plt.subplots(figsize=(10,10))
sns.heatmap(datac,linewidth=.2,annot=True)
```

There is no highly corelated features in the dataset.

# a. ph and Chloramines

```python
df[['ph','Chloramines']].corr()
```

```python
sns.scatterplot(data =df, x = 'Chloramines', y = 'ph')
plt.title('Chloramines vs ph')
plt.show()
```

Corelation value is -.024768,which is approximately equal to zero.We can conclude that,there is no relevant corelation between ph and Chloramines

# b. ph and Conductivity 

```python
df[['ph','Conductivity']].corr()
```

```python
sns.scatterplot(data =df, x = 'Conductivity', y = 'ph')
plt.title('Conductivity vs ph')
plt.show()
```

Corelation value is .014128 ,which is approximately equal to zero.We can conclude that,there is no relevant corelation between ph and Conductivity.
