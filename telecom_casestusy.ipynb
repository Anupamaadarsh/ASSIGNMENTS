```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
df=pd.read_excel('C:\\Users\\HP\\Documents\\week4\\Telco-Customer-Churn.xlsx')
plt.figure(figsize = (14,5))

plt.subplot(1,3,1)
sns.histplot(y= 'PaymentMethod', data =df, hue = 'gender')
plt.ylabel("")
plt.title("Payment methods across gender")

plt.subplot(1,3,2)
sns.histplot(y= 'PaymentMethod', data =df, hue = 'SeniorCitizen')
plt.title("Payment methods across Citizen ")
plt.ylabel("")

plt.subplot(1,3,3)
sns.histplot(y = 'PaymentMethod', data = df, hue ='Churn')
plt.title("Payment methods & Churn")
plt.ylabel("")
```
